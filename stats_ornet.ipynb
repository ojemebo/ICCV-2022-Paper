```python
# calculate number of objects in image

# first read the image and convert to grayscale
image = cv2.imread('C:/Users/blessing/Desktop/binary.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Next, blur the image to avoid noises using any of the filters
blur = cv2.GaussianBlur(gray, (11, 11), 0)

# detect edges using a canny algorithm
canny = cv2.Canny(blur, 30, 150, 3)

# connect the edges to make them thiker and more visible
dilated = cv2.dilate(canny, (1, 1), iterations=0)

#  calculate the contour in image and convert image into RGB from BGR and then draw the contours
(cnt, hierarchy) = cv2.findContours(
    dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)
 
 
print("The number of objects in this image: ", len(cnt))
```

```python
filename = 'C:/Users/blessing/Desktop/fine_seg_control.png'
img = cv2.imread(filename)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# find Harris corners
gray = np.float32(gray)
dst = cv2.cornerHarris(gray,2,3,0.04)
dst = cv2.dilate(dst,None)
ret, dst = cv2.threshold(dst,0.01*dst.max(),255,0)
dst = np.uint8(dst)

# find centroids
ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)

# define the criteria to stop and refine the corners
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.001)
corners = cv2.cornerSubPix(gray,np.float32(centroids),(5,5),(-1,-1),criteria)

# Now draw them
res = np.hstack((centroids,corners))
res = np.int0(res)
img[res[:,1],res[:,0]]=[0,0,255]
img[res[:,3],res[:,2]] = [0,255,0]
plt.imshow(corners)
cv2.imwrite('subpixel5.png',img)
```
